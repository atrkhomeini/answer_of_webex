```python
%%rdf turtle -p
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
```

CREATE NEW CLASS

```python
%%rdf turtle -l sentiment
@prefix bp: <http://example.org/brand-perception#> .
@prefix onyx: <http://www.gsi.dit.upm.es/ontologies/onyx/ns#> .
@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .

# Membuat subclass untuk sentimen yang lebih spesifik
bp:HighlyPositiveSentiment a rdfs:Class ;
    rdfs:label "Highly Positive Sentiment" ;
    rdfs:comment "Sentimen sangat positif yang menunjukkan apresiasi tinggi terhadap merek." ;
    rdfs:subClassOf onyx:Sentiment .

bp:NegativeSentiment a rdfs:Class ;
    rdfs:label "Negative Sentiment" ;
    rdfs:comment "Sentimen negatif yang menunjukkan ketidakpuasan terhadap merek." ;
    rdfs:subClassOf onyx:Sentiment .

bp:MixedSentiment a rdfs:Class ;
    rdfs:label "Mixed Sentiment" ;
    rdfs:comment "Sentimen campuran yang mencakup baik aspek positif maupun negatif terhadap merek." ;
    rdfs:subClassOf onyx:Sentiment .
```

CREATE NEW PROPERTIES

```python
%%rdf turtle -l sentiment
@prefix bp: <http://example.org/brand-perception#> .
@prefix onyx: <http://www.gsi.dit.upm.es/ontologies/onyx/ns#> .
@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .
# Properti untuk mengukur intensitas sentimen
bp:sentimentIntensity a rdf:Property ;
    rdfs:label "Sentiment Intensity" ;
    rdfs:comment "Tingkat intensitas sentimen, dari 0 (lemah) hingga 1 (kuat)." ;
    rdfs:domain onyx:Sentiment ;
    rdfs:range xsd:float .

# Properti untuk menunjukkan kategori sentimen
bp:sentimentCategory a rdf:Property ;
    rdfs:label "Sentiment Category" ;
    rdfs:comment "Kategori sentimen spesifik yang berhubungan dengan persepsi merek." ;
    rdfs:domain onyx:Sentiment ;
    rdfs:range rdfs:Literal .
```
